Исходные данные имеют следующую структуру:

```

directories = {
```

````python
# Data
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
````

````python
# 1.1 Поиск имени по номеру документа
def search_name(doc_no):
    """
    Ищет имя по номеру документа
    """
    # doc_no = str(input('Введите номер документа:'))
    for docs in documents:
        if docs['number'] == doc_no:
            return docs['name']
        else:
            None
    # return 'Документ не найден в базе'
````

````python
# 1.2 Поиск полки по номеру документа
def search_doc(doc_no):
    """
    Ищет полку документа по его номеру
    """    
    # doc_no = str(input('Введите номер документа:'))
    for k, v in directories.items():
        if doc_no in v:
            return k
        else:
            None
    # return
````

````python
# 1.3 Информация
def list_docs():
    """
    Показывет информацию по всем документам
    """
    list_docs = []
    for docs in documents:
        n = docs['number']
        t = docs['type']
        na = docs['name']
        s = search_doc(docs['number'])
        print(f'№: {n}, тип: {t}, владелец: {na}, полка хранения: {s}')
    return
    
````

````python
# 1.4 Добавление полки
def shelf_add(shelf_no):
    """
    Добавляет новую полку
    """
    if shelf_no not in directories.keys():
        return directories.setdefault(shelf_no, [])
````

````python
# 1.5 Удаление полки
def shelf_del(shelf_no):
    """
    Удаляет полку
    """
    if shelf_no in directories.keys():
        del(directories[shelf_no])
````

````python
# Проверка полок
def shelf_check(shelf_no):
    """
    Ищет наличие полки
    """
    if shelf_no in directories.keys():
        return True
    return False
````

````python
# Список полок
def shelf_list():
    """
    отображает текущий список полок
    """    
    sl = ', '.join(str(x) for x in list(directories.keys()))
    return sl
````

````python
# Содержимое полки
def shelf_contents(shelf_no):
    """
    Проверяет наличие документов на полке
    """
    return len(directories[shelf_no])
````

````python
# запрос данных о документе
def doc_ask():
    """
    Запрашивает данные о документе для операций создания, удаления и перемещения
    """
    doc_no = str(input('Введите номер документа:'))
    doc_type = str(input('Введите тип документа:'))
    doc_name = str(input('Введите владельца документа:'))
    shelf_no = str(input('Введите полку для хранения:'))
    doc_ask_dict = {'doc_no': doc_no, 'doc_type': doc_type, 'doc_name': doc_name, 'shelf_no': shelf_no}
    return doc_ask_dict
````

````python
# поиск документа по номеру
def doc_search(doc_no):
    """
    Ищет индекс документа в списке по номеру
    """
    for i, lines in enumerate(documents):
        if lines['number'] == doc_no:
            return i
    return False
````

````python
# Создание документа на полке
def doc_add(doc_ask_dict):
    """
    Создаёт новый документ и запись на соответствующей полке
    """
    directories[doc_ask_dict['shelf_no']].append(doc_ask_dict['doc_no']) # создаёт ID документа на полке
    documents.append({'type': doc_ask_dict['doc_type'], 'number': doc_ask_dict['doc_no'], 'name': doc_ask_dict['doc_name']}) # создаёт документ
````

````python
# Удаление документа
def doc_del(doc_no):
    """
    Удаляет документ и запись на соответствующей полке
    """
    documents.pop(doc_search(doc_no))
    return directories[str(search_doc(doc_no))].remove(doc_no)
````

````python
# Перемещение документа
def doc_move(shelf_no, doc_no):
    """
    Перекладываент документ на другую полку
    """
    directories[search_doc(doc_no)].remove(doc_no)
    return directories[shelf_no].append(doc_no)
````

````python
# Main
def main():
    while prog() != 'q':
        prog()
        continue
    return 'System shutdown'
````

````python
# Выбор программы
# можно было сообщения записать в функции, но мне показалось так менее удобно
# функции выполняют какие-то действия, а все сообщения прописаны в отдельном окне
# тем более, что так проще корректировать текст, если нужно, чтобы для разных команд
# текст немного отличался
def prog():
    cmd = input('Введите команду:')
    if cmd == 'q': # прекращение работы
        return 'q'

    elif cmd == 'p': # показать имя владельца по номеру
        doc_no = input('Введите номер документа:')
        if search_name(doc_no) == None:
            return print('Документ не наден в базе')
        else:
            return print(f'Владелец документа: {search_name(doc_no)}')

    elif cmd == 's': # показать полку хранения по номеру документа
        doc_no = input('Введите номер документа:')
        if search_doc(doc_no) == None:
            return print('Документ не наден в базе')
        else:
            return print(f'Документ хранится на полке: {search_doc(doc_no)}')

    elif cmd == 'l': # список всех документов
        return list_docs()

    elif cmd == 'ads': # добавление полки
        shelf_no = str(input('Введите номер полки:'))
        if shelf_check(shelf_no) is False:
            shelf_add(shelf_no)
            return print (f"Полка добавлена. Текущий перечень полок:", shelf_list())
        return print(f"Такая полка уже существует. Текущий перечень полок:", shelf_list())

    elif cmd == 'ds': # удаление полки
        shelf_no = str(input('Введите номер полки:'))
        if shelf_check(shelf_no) is False:
            return print (f"Такой полки не существует. Текущий перечень полок:", shelf_list())
        elif shelf_check(shelf_no) is True:
            if shelf_contents(shelf_no) != 0:
                return print (f"На полке есть документы. Текущий перечень полок:", shelf_list())
            elif shelf_contents(shelf_no) == 0:
                shelf_del(shelf_no)    
                return print(f"Полка удалена. Текущий перечень полок:", shelf_list())
        return print('Что-то странное')

    elif cmd == 'ad': # добавление документа
        doc_ask_dict = doc_ask()
        if shelf_check(doc_ask_dict['shelf_no']) is False:
            print('Такой полки не существует. Добавьте полку командой as.\nТекущий список документов:')
            return list_docs()
        elif shelf_check(doc_ask_dict['shelf_no']) is True:
            doc_add(doc_ask_dict)
            print('Документ добавлен. Текущий список документов:')
            return list_docs()   

    elif cmd == 'd': # удаление документа
        doc_no = str(input('Введите номер документа:'))
        if doc_search(doc_no) is False:
            print('Документ не найден в базе.\nТекущий список документов:')
            return list_docs()
        elif doc_search(doc_no) is not False:
            doc_del(doc_no)
            print('Документ удалён.\nТекущий список документов:')
            return list_docs()

    elif cmd == 'm': # перемещение документа
        doc_no = str(input('Введите номер документа:'))
        shelf_no = str(input('Введите номер полки:'))
        if doc_search(doc_no) is False:
            print('Документ не найден в базе.\nТекущий список документов:')
            return list_docs()
        elif shelf_check(shelf_no) is False:
            return print (f"Такой полки не существует. Текущий перечень полок:", shelf_list())
        elif doc_search(doc_no) is not False:
            doc_move(shelf_no, doc_no)
            print('Документ перемещён. Текущий список документов:')
            return list_docs()
            
    else:
        return print('Некорректная команда')
````
